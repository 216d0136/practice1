```


from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()
x = iris.data[:, 2:] # 花びらの長さと幅
y = iris.target
model = DecisionTreeClassifier().fit(x,y)

"""
criterion・・・　分類基準（giniとentropyがある）
不純度には二つの指標があり、
一つ目はジニ不純度(gini)
二つ目が情報エントロピー(entopy)。
不純度は、ひとつのノード内に複数のクラスがどれだけ混在(混ざっているか)しているかを表す指標。
max_depth・・・　木の深さ
min_samples_leaf・・・　分割するときに必要なデータ数
"""

import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
%matplotlib inline

fig = plt.figure(figsize=(15, 8))
ax = fig.add_subplot()
plot_tree(model, feature_names=iris.data[:, 2:] , ax=ax, filled=True, class_names=['setosa', 'versicolor', 'virginica']);
```
